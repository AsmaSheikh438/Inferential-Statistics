# Hypothesis
Hypothesis is define as making a guess (not wild guess) based on assumption without scientific proof or explaining the situation based on reasonable assumptions.

# Hypothesis Testing
* It is a part of statical analysis, where we test the Assumptions made regarding a population parameter.
* It is generally used when we were to compare a single group with an external standard and two or more groups with each other.
* we genrally work on sample data in hypothesis testing.

### Null Hypothesis
* Null hypothesis is a statical theory that suggest there is no statical significance exist bewteen the population .
* There is no significant difference between the two samples.
* It is denoted by H0 and read as H-naught
##### Null Hypothesis H0: 
No effect exists in the population.

### Alternative Hypothesis
* An Alternative Hypothesis suggests there is a significant difference between the population parameters. 
* It could be Greater or Smaller.
* Basically, it is the contrast of the null hypothesis.
* It is denoted by Ha or H1.

##### Alternative Hypothesis Ha or H1
The effect exists in the population.
