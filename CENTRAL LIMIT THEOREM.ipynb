 ### Central Limit Theorem works in both 
 * Normal Distribution
 * Non-Normal Distribution

 * (CLT) use when we have population data in very large amount

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# generate randombly data

pop_data = [np.random.randint(10,100) for i in range(10000)]
len(pop_data)
```

```python
# creating DataFrame of 10,000 data....

pop_table = pd.DataFrame({'pop_data)':pop_data})
pop_table
```

```python
# checking distribution of data

plt.figure(figsize=(5,4))
sns.kdeplot(x="pop_data)",data = pop_table)
plt.show()
```

```python
# The above data shows in graph is not in Normal Distribution (non-normal distribution)
#  Now we will collect some sample from population data and we will find the out the sample data would be in normal distribution or not.
# We can only pick up the 10% of data from the population and collect more than 30 samples of data , 10% more than 30 sample
```

```python
# checking 10% of data
(10000*10)/100
```

```python
# now picking sample data and save its mean
sample_mean = []
for no_sample in range(50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
    
```

```python
sample_M = pd.DataFrame({'Sample_Mean': sample_mean})
```

```python
plt.figure(figsize=(5,4))
sns.kdeplot(x='Sample_Mean',data = sample_M)
plt.show()
```

```python
# now this graph showing Normal Distribution data so, that was how we can convert POPULATION DATA into NORMAL DISTRIBUTION DATA
```

```python
# we can change the range of sample data for range (60)

sample_mean = []
for no_sample in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
    
```

```python
plt.figure(figsize=(5,4))
sns.kdeplot(x='Sample_Mean',data = sample_M)
plt.show()
```

#### The most important line of a central limit theorem is population data mean is almost same to sample data mean of (CLT) central limit theorem.

```python
# calculating mean of population data

np.mean(pop_data)
```

```python
# calculating mean of sample data

np.mean(sample_mean)
```

#### We can  see clearly the mean of population data and mean of sample data is almost same
